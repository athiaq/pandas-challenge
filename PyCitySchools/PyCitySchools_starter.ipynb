# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_school = school_data_complete['school_name'].unique()
total_school_number = len(unique_school)
total_school_number
```

```python
# Calculate the total number of students
total_students = school_data_complete['student_name'].count()
total_students
```

```python
# Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete['math_score'] >= 70)].count()['student_name']
passing_math_percentage = passing_math_count / float(total_students) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete['reading_score'] >=70)].count()['student_name']
passing_reading_percentage = passing_reading_count / float(total_students) *100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(total_students) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({'Total Schools':[total_school_number], 'Total Students': [total_students], 'Total Budget': [total_budget], 
                                'Average Math Score': [average_math_score], 'Average Reading Score': [average_reading_score], 'Passing Math Percentage': [passing_math_percentage],
                                'Passing Reading Percentage': [passing_reading_percentage], 'Passing Math and Reading Percentage': [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type
school_types = school_data_complete.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
per_school_counts = school_data_complete['school_name'].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_student_budget = per_school_budget/per_school_counts
display(per_school_budget, per_student_budget)
```

```python
# Calculate the average test scores for math per school
average_math_score = school_data_complete.groupby(['school_name']).mean()['math_score']
# Calculate the average test scores for reading per school
average_reading_score = school_data_complete.groupby(['school_name']).mean()['reading_score']
#Calculate the average test scores for math and reading combined per school 
average_scores_math_reading = (average_reading_score + average_math_score)/2
display(average_math_score, average_reading_score, average_scores_math_reading)
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = school_data_complete[(school_data_complete['math_score'] >= 70)]
school_passing_math.head()
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = school_data_complete[(school_data_complete['reading_score'] >= 70)]
school_passing_reading.head()
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"]/per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"]/per_school_counts * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"]/per_school_counts * 100
display(overall_passing_rate)
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.DataFrame(
    {
        "School Type": school_types,
        "Total Students": per_school_counts,
        "Total School Budget": per_school_budget,
        "Per Student Budget": per_student_budget,
        "Average Math Score": average_math_score,
        "Average Reading Score": average_reading_score,
        "% Passing Math": per_school_passing_math,
        "% Passing Reading": per_school_passing_reading,
        "% Overall Passing": overall_passing_rate
    }
)

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"]
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"]

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
school_spending_df.info()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Student Spending"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels = labels, include_lowest = True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Student Spending"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Student Spending"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Student Spending"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Student Spending"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Student Spending"]).mean()["% Overall Passing"]
display(overall_passing_spending, spending_math_scores, spending_reading_scores, spending_passing_reading, spending_passing_math)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Overall Spending": overall_passing_spending,
    "Spending Math Scores": spending_math_scores,
    "Spending Reading Scores": spending_reading_scores,
    "Spending Passing Reading": spending_passing_reading,
    "Spending Passing Math": spending_passing_math
})

# Display results
spending_summary
```
